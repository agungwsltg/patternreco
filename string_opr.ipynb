```python
# Metoda Teks: upper => Membuat jadi Huruf BESAR 
txt = "Pengenalan Pola"
print(txt.upper())
```

```python
# Metoda Teks: lower => Membuat jadi Huruf kecil 
txt = "Pengenalan Pola"
print(txt.lower())
```

```python
# Metoda Teks: title => Huruf Besar di setiap awal Kata 
txt = "pengenalan pola"
print(txt.title())
```

```python
# Metoda Teks: swapcase => Mengubah Huruf besar menjadi kecil dan
# mengubah huruf kecil menjadi Besar 
txt = "Pengenalan Pola"
print(txt.swapcase())
```

```python
# Metoda Teks: strip => menghilangkan spasi di kiri dan kanan
txt1 ="             Belajar           "
txt2 = "Pengenalan Pola"
print(txt1.strip() + txt2)
```

```python
# Metoda Teks: strip => menghilangkan spasi di kiri dan kanan
txt1 ="             Belajar           "
txt2 = "Pengenalan Pola"
print(txt1.lstrip() + txt2)
```

```python
# Metoda Teks: strip => menghilangkan spasi di kiri dan kanan
txt1 ="             Belajar           "
txt2 = "Pengenalan Pola"
print(txt1.rstrip() + txt2)
```

```python
# Metoda Teks: split => memecah kalimat menjadi kata 
txt = "Belajar Pengenalan Pola"
print(txt.split())
```

```python
# Metoda Teks: translate => mengganti huruf dengan huruf lainnya
# mengganti huruf P dengan F 
txt = "Belajar Pengenalan Pola"
txtubah = txt.maketrans('P','F') 
print(txt.translate(txtubah))
```

```python
# Tuple: MengUpdate isi Tuple Alpukat diubah menjadi Jeruk Bali
# Isi Tuple tidak dapat diubah, untuk  mengubah isi tuple
# Ubah dulu menjadi list
buah = ("Mangga", "Nanas", "Jambu", "alpukat", "Durian")
ubah= list(buah)
ubah[3] = "Jeruk Bali"
buah = tuple(ubah)

print(buah)
```

```python
# Tuple: Menambah isi Tuple dengan Jeruk Bali
# Isi Tuple tidak dapat ditambah, untuk  menambah isi tuple
# Ubah dulu menjadi list
buah = ("Mangga", "Nanas", "Jambu", "alpukat", "Durian")
ubah= list(buah)
ubah.append("Jeruk Bali")
buah = tuple(ubah)

print(buah)
```

```python
# Set merupakan salah satu dari empat type data bawaan  python tidak terurut 
# dan tidak terinde yang digunakan untk menyimpan kumpulan data. Isi Set tidak 
# dapat diubah tidak diperkenan duplikat data

# Menampilkan Isi Set

hari = {"Senin", "Selasa", "Rabu", "Kamis", "Jum'at", "Sabtu", "Minggu"}
print(hari)
```

```python
# Apabila ada isi data yang sama dalam Set, maka akan di tampilkan satu

hari = {"Senin", "Selasa", "Rabu", "Kamis", "Jum'at", "Sabtu", "Minggu", "Senin"}
print(hari)
```

```python
# Set: Menambah isi dari Set
buah = {"Mangga", "Nanas", "Jambu", "alpukat", "Durian"}
buah.add("Jeruk Bali")

print(buah)
```

```python
# Set: Menambah isi dari Set dengan Set
buah = {"Mangga", "Nanas", "Jambu", "alpukat", "Durian"}
buahnambah = {"Jeruk Bali","Melon", "Lemon"}
buah.update(buahnambah)
print(buah)
```

```python
# Set: Menghapus isi dari Set dengan Remove
buah = {"Mangga", "Nanas", "Jambu", "alpukat", "Durian"}
buah.remove("Nanas")

print(buah)
```

```python
# Set: Menghapus isi dari Set dengan Discard
buah = {"Mangga", "Nanas", "Jambu", "alpukat", "Durian"}
buah.discard("Nanas")

print(buah)
```

```python
# Set: Menampilkan isi Set dengan Looping
hari = {"Senin", "Selasa", "Rabu", "Kamis", "Jum'at", "Sabtu", "Minggu"}
for x in hari:
  print(x)
```

```python
# Dictionaty / Kamus: menampilkan isi data dari jenis

kamus = {
  "jenis": "Arial",
  "warna": "Hitam",
  "ukuran": 12,
  "style": "Italic"
}
print(kamus["jenis"])
```

```python
# Dictionaty / Kamus: menampilkan isi data dari jenis 
# menggunakan get()

kamus = {
  "jenis": "Arial",
  "warna": "Hitam",
  "ukuran": 12,
  "style": "Italic"
}
print(kamus.get("jenis"))
```

```python
# Dictionaty / Kamus: menampilkan atribut/ kunci dari Kamus 

kamus = {
  "jenis": "Arial",
  "warna": "Hitam",
  "ukuran": 12,
  "style": "Italic"
}
print(kamus.keys())
```

```python
# Dictionaty / Kamus: menambah isi dari Kamus 

kamus = {
  "jenis": "Arial",
  "warna": "Hitam",
  "ukuran": 12,
  "style": "Italic"
}
kamus["underline"] = True 
print(kamus)
```
